```python
import csv
import numpy as np
import pandas as pd
import json
```

```python
KDs = pd.read_csv("Data/CFSKDsForVisualization.csv")
Modules = pd.read_csv("Data/CFSModulesForVisualization.csv")
Network = pd.read_csv("Data/CFSNetworkForVisualization_withoutNaN.csv")
```

### Modules of interest

1) NKM4

2) BcellM7

3) MonoM6

4) NKM13

5) NKM9

```python
Modules
```

```python
NKM4 = []
BcellM7 = []
MonoM6 = []
NKM13 = []
NKM9 = []

for i in range(len(Modules)):
    if Modules["Module"][i] == "NKM4":
        NKM4.append(Modules["GeneName"][i])
    if Modules["Module"][i] == "BcellM7":
        BcellM7.append(Modules["GeneName"][i])
    if Modules["Module"][i] == "MonoM6":
        MonoM6.append(Modules["GeneName"][i])
    if Modules["Module"][i] == "NKM13":
        NKM13.append(Modules["GeneName"][i])
    if Modules["Module"][i] == "NKM9":
        NKM9.append(Modules["GeneName"][i])  
```

```python
print(len(NKM4), len(BcellM7), len(MonoM6), len(NKM13), len(NKM9))
```

## Nodes creation

```python
parents = np.unique(list(Network["ParentGeneName"]))
children = np.unique(list(Network["ChildGeneName"]))

nodes = list(parents) + list(children)
nodes = np.unique(nodes)
```

```python
nodes_and_groups = []

for i in nodes:
    if i in NKM4:
        nodes_and_groups.append([i, "NKM4"])
    if i in BcellM7:
        nodes_and_groups.append([i, "BcellM7"])
    if i in MonoM6:
        nodes_and_groups.append([i, "MonoM6"])
    if i in NKM13:
        nodes_and_groups.append([i, "NKM13"])
    if i in NKM9:
        nodes_and_groups.append([i, "NKM9"])
    else:
        nodes_and_groups.append([i, "none"])
        
```

## Links creation

```python
links = []
for parent, child in zip(Network["ParentGeneName"], Network["ChildGeneName"]):
    links.append([parent, child])
```

```python
final = {"nodes": [], "links": []}

for parent, group in parents_and_groups:
    final["nodes"] += [{"id": parent, "group": group}]

for child, group in children_and_groups:
    final["nodes"] += [{"id": child, "group": group}]

for parent, child in links:
    final["links"] += [{"source": parent, "target": child, "value": 1}]
```

```python
with open('complete_data.json', 'w') as fp:
    json.dump(final, fp)
```

```python
## JOINING PARENTS AND CHILD TO NODES

nodes = []

for i in parents:
    nodes.append(i)
for i in children:
    nodes.append(i)
    
nodes = np.unique(nodes)
```

```python
## LINKS OF INTEREST - ADD TO THE LIST IF EITHER PARENT OR CHILD ARE IN THE DESIRED MODULE

links_BcellM7 = []

for i in nodes:
    if i in BcellM7:
        for p, c in links:
            if p == i:
                links_BcellM7.append([p, c])
            if c == i:
                links_BcellM7.append([p, c])
                       
```

```python
## NODES OF INTEREST

nodes_BcellM7 = np.unique(links_BcellM7)
```

```python
modules_nodes = {}
for i in range(len(Modules)):
    modules_nodes[Modules["GeneName"][i]] = Modules["Module"][i]
    
BcellM7_KD = {}
for kd, gene in zip(KDs["BcellM7_KD"], KDs["geneName_KD"]):
    BcellM7_KD[gene] = kd
```

```python
nodes_info = {}

for node in nodes_BcellM7:
    if node in modules_nodes and node in BcellM7_KD:
        nodes_info[node] = [modules_nodes[node], BcellM7_KD[node]]
    if node in modules_nodes and node not in BcellM7_KD:
        nodes_info[node] = [modules_nodes[node], "None"]
    if node not in modules_nodes and node in BcellM7_KD:
        nodes_info[node] = ["None", BcellM7_KD[node]]
    if node not in modules_nodes and node not in BcellM7_KD:
        nodes_info[node] = ["None", "None"]
```
